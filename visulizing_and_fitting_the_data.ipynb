```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./pot_data_with_motor_s_1.csv')
```

```python
df.columns
```

```python
import numpy as np
```

```python
plt.figure(figsize=(10, 5))
plt.plot(np.array(df['Potentiometer_Data']), np.array(df['Motor_Position']), marker='o', linestyle='-')

# Customize the plot
plt.title('Motor Position vs Potentiometer Data')
plt.xlabel('Potentiometer Data')
plt.ylabel('Motor Position')
plt.grid(True)

# Show the plot
plt.show()
```
